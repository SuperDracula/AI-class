```python
import random
```

```python
def get_actions(matrix):
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                zero_pos = (i, j)
                break
    r, c = zero_pos

    possible_moves = []
    if r > 0: possible_moves.append("UP")
    if r < 2: possible_moves.append("DOWN")
    if c > 0: possible_moves.append("LEFT")
    if c < 2: possible_moves.append("RIGHT")
    return r, c, possible_moves
```

```python
def rule_match(pos_moves, last_move):
    xdirections = {"UP": "DOWN", "DOWN": "UP", "LEFT": "RIGHT", "RIGHT": "LEFT"}
    return random.choice([a for a in pos_moves if a != xdirections.get(last_move, "")])
```

```python
def main():
    matrix =   [[1, 5, 6],
                [2, 0, 7],
                [3, 4, 8]]
    steps = 0
    last_move = None
    max_steps = 1000
    while steps < max_steps:
        steps += 1
        r, c, actions = get_actions(matrix)
        action = rule_match(actions, last_move)
        print(f"Step {steps + 1}: Action: {action}")
        last_move = action

        if action == "UP":
            matrix[r][c], matrix[r - 1][c] = matrix[r - 1][c], matrix[r][c]
        elif action == "DOWN":
            matrix[r][c], matrix[r + 1][c] = matrix[r + 1][c], matrix[r][c]
        elif action == "LEFT":
            matrix[r][c], matrix[r][c - 1] = matrix[r][c - 1], matrix[r][c]
        elif action == "RIGHT":
            matrix[r][c], matrix[r][c + 1] = matrix[r][c + 1], matrix[r][c]

        if matrix == [[1, 2, 3], [4, 0, 5], [6, 7, 8]]:
            print("Puzzle Solved!")
            break

        print(f"Current State: \n{matrix}\n")

if __name__ == "__main__":
    main()
```
