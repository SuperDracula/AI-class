```python
import random
import queue
import pygame
import sys
import time
import math

GOAL_MATRICES = []

def apply_move(matrix, move):
    r, c = -1, -1
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                r, c = i, j
                
    new_matrix = [row[:] for row in matrix]
    if move == "UP" and r > 0:
        new_matrix[r][c], new_matrix[r - 1][c] = new_matrix[r - 1][c], new_matrix[r][c]
    elif move == "DOWN" and r < 2:
        new_matrix[r][c], new_matrix[r + 1][c] = new_matrix[r + 1][c], new_matrix[r][c]
    elif move == "LEFT" and c > 0:
        new_matrix[r][c], new_matrix[r][c - 1] = new_matrix[r][c - 1], new_matrix[r][c]
    elif move == "RIGHT" and c < 2:
        new_matrix[r][c], new_matrix[r][c + 1] = new_matrix[r][c + 1], new_matrix[r][c]
        
    return new_matrix

def generate_2_states(base_matrix):
    """Sinh ra 2 ma trận hợp lệ từ 1 ma trận khuyết (-1)"""
    coords = []
    for r in range(3):
        for c in range(3):
            if base_matrix[r][c] == -1:
                coords.append((r, c))
                
    m1 = [row[:] for row in base_matrix]
    m2 = [row[:] for row in base_matrix]
    
    # Ma trận 1: Để 8 vào ô khuyết 1, 0 vào ô khuyết 2
    m1[coords[0][0]][coords[0][1]] = 8
    m1[coords[1][0]][coords[1][1]] = 0
    
    # Ma trận 2: Để 0 vào ô khuyết 1, 8 vào ô khuyết 2
    m2[coords[0][0]][coords[0][1]] = 0
    m2[coords[1][0]][coords[1][1]] = 8
    
    return [m1, m2]

def get_state(cost, node_id, state_matrices, parent_node, move, step):
    return [cost, node_id, state_matrices, parent_node, move, step]

def to_tuple(state_matrices):
    return tuple(tuple(tuple(row) for row in m) for m in state_matrices)

def cal_ucs_cost(parent_node_state):
    if parent_node_state is None:
        return 0
    return parent_node_state[0] + 1

def manhattan(m, goal_matrix):
    goal_flat = [e for sublist in goal_matrix for e in sublist]
    diff = 0
    for i in range(3):
        for j in range(3):
            val = m[i][j]
            if val != 0:
                for gi in range(3):
                    for gj in range(3):
                        if goal_matrix[gi][gj] == val:
                            diff += abs(gi - i) + abs(gj - j)
    return diff

def cal_cost(child_matrices):
    total = 0
    for m in child_matrices:
        total += min(manhattan(m, g) for g in GOAL_MATRICES)
    return total / len(child_matrices)

def is_goal(state_matrices):
    for m in state_matrices:
        if m not in GOAL_MATRICES:
            return False
    return True

def expand(state_matrices):
    children = []
    moves = ["UP", "DOWN", "LEFT", "RIGHT"]
    random.shuffle(moves)
    
    for move in moves:
        new_matrices = []
        changed = False
        
        for m in state_matrices:
            new_m = apply_move(m, move)
            if new_m != m:
                changed = True
            new_matrices.append(new_m)
            
        if changed:
            children.append((new_matrices, move))
            
    return children

def extract_path(final_node):
    path = []
    current = final_node
    while current is not None:
        matrices_flat = []
        for m in current[2]:
            matrices_flat.append([val for row in m for val in row])
        move = current[4] if current[4] is not None else "START" 
        path.append({'matrix': matrices_flat, 'move': move})
        current = current[3] 
    path.reverse() 
    return path

def do_BFS(reached, frontier, frontier_set):
    step = 0
    while not frontier.empty() and step < 281000:
        node_state = frontier.get()
        node = node_state[2] 
        if is_goal(node):
            return node_state 
        reached.add(to_tuple(node))
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            s = get_state(0, 0, child, node_state, move, step)
            if is_goal(child):
                return s 
            if child_tuple not in reached and child_tuple not in frontier_set:
                frontier.put(s)
                frontier_set.add(child_tuple)
    return None

def do_DFS(reached, frontier, frontier_set):
    step = 0
    while frontier and step < 281000:
        node_state = frontier.pop()
        node = node_state[2] 
        if is_goal(node):
            return node_state 
        reached.add(to_tuple(node))
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            s = get_state(0, 0, child, node_state, move, step)
            if is_goal(child):
                return s 
            if child_tuple not in reached and child_tuple not in frontier_set:
                frontier.append(s)
                frontier_set.add(child_tuple)
    return None

def do_DLS(reached, frontier, frontier_set, limit_depth):
    step = 0
    while frontier and step < 281000:
        node_state = frontier.pop()
        node = node_state[2] 
        if is_goal(node):
            return node_state 
        if node_state[5] >= limit_depth:
            continue
        reached.add(to_tuple(node))
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            s = get_state(0, 0, child, node_state, move, step) 
            if is_goal(child):
                return s
            if child_tuple not in reached and child_tuple not in frontier_set:
                frontier.append(s)
                frontier_set.add(child_tuple)
                reached.add(child_tuple)
    return None

def do_UCS(reached, frontier):
    step = 0
    node_id = 0
    while not frontier.empty() and step < 281000:
        node_state = frontier.get()
        node = node_state[2]
        node_cost = node_state[0]
        if is_goal(node):
            return node_state 
        node_tuple = to_tuple(node)
        if node_tuple in reached and reached[node_tuple] < node_cost:
            continue
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            cost = cal_ucs_cost(node_state)
            if child_tuple not in reached or cost < reached[child_tuple]:
                node_id += 1
                s = get_state(cost, node_id, child, node_state, move, step) 
                frontier.put(s)
                reached[child_tuple] = cost
    return None

def do_Greedy(reached, frontier):
    step = 0
    node_id = 0
    while not frontier.empty() and step < 281000:
        node_state = frontier.get()
        node = node_state[2]
        node_cost = node_state[0]
        if is_goal(node):
            return node_state 
        node_tuple = to_tuple(node)
        if node_tuple in reached and reached[node_tuple] < node_cost:
            continue
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            cost = cal_cost(child)
            if child_tuple not in reached or cost < reached[child_tuple]:
                node_id += 1
                s = get_state(cost, node_id, child, node_state, move, step) 
                frontier.put(s)
                reached[child_tuple] = cost
    return None

def do_AStar(reached, frontier):
    step = 0
    node_id = 0
    while not frontier.empty() and step < 281000:
        node_state = frontier.get() 
        node = node_state[2]
        node_g = node_state[5]
        if is_goal(node):
            return node_state 
        node_tuple = to_tuple(node)
        if node_tuple in reached and reached[node_tuple] < node_g:
            continue
        reached[node_tuple] = node_g
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            g_n = node_g + 1
            h_n = cal_cost(child)
            f_n = g_n + h_n
            if child_tuple not in reached or g_n < reached.get(child_tuple, float('inf')):
                node_id += 1
                s = get_state(f_n, node_id, child, node_state, move, g_n)
                frontier.put(s) 
    return None

def do_IDAStar(reached, frontier, limit_f):
    step = 0
    node_id = 0
    next_limit = float('inf') 
    while frontier and step < 281000:
        node_state = frontier.pop() 
        node = node_state[2]
        node_g = node_state[5]
        if is_goal(node):
            return node_state, next_limit 
        node_tuple = to_tuple(node)
        if node_tuple in reached and reached[node_tuple] < node_g:
            continue
        reached[node_tuple] = node_g
        child_move = expand(node)
        step += 1
        for child, move in child_move:
            child_tuple = to_tuple(child)
            g_n = node_g + 1
            h_n = cal_cost(child)
            f_n = g_n + h_n
            if f_n <= limit_f:
                if child_tuple not in reached or g_n < reached.get(child_tuple, float('inf')):
                    node_id += 1
                    s = get_state(f_n, node_id, child, node_state, move, g_n)
                    frontier.append(s)
            else:
                next_limit = min(next_limit, f_n)
    return None, next_limit

def do_SimpleHillClimbing(start_state):
    start_tuple = to_tuple(start_state)
    current_h = cal_cost(start_state)
    current_state = get_state(current_h, 0, start_state, None, "START", 0)
    step = 0 
    while step < 281000:
        node = current_state[2]
        current_h = current_state[0]
        if is_goal(node):
            return current_state
        child_move = expand(node)
        found_better = False
        step += 1
        for child, move in child_move:
            child_h = cal_cost(child)
            if child_h < current_h:
                current_state = get_state(child_h, 0, child, current_state, move, step)
                found_better = True
                break 
        if not found_better:
            print("Đã bị kẹt ở đỉnh cục bộ! Thuật toán SHC dừng lại.")
            return current_state
    return None

def do_SteepestAscent(start_state):
    start_tuple = to_tuple(start_state)
    current_h = cal_cost(start_state)
    current_state = get_state(current_h, 0, start_state, None, "START", 0)
    step = 0 
    while step < 281000:
        node = current_state[2]
        current_h = current_state[0]
        if is_goal(node):
            return current_state
        child_move = expand(node)
        step += 1
        best_child = None
        best_h = float('inf')
        best_move = None
        for child, move in child_move:
            child_h = cal_cost(child)
            if child_h < best_h:
                best_h = child_h
                best_child = child
                best_move = move
        if best_h < current_h:
            current_state = get_state(best_h, 0, best_child, current_state, best_move, step)
        else:
            print("Đã bị kẹt ở đỉnh cục bộ! Thuật toán Steepest-Ascent dừng lại.")
            return current_state
    return None

def do_StochasticHillClimbing(start_state):
    current_h = cal_cost(start_state)
    current_state = get_state(current_h, 0, start_state, None, "START", 0)
    step = 0 
    while step < 281000:
        node = current_state[2]
        current_h = current_state[0]
        if is_goal(node):
            return current_state
        child_move = expand(node)
        step += 1
        better_moves = []
        for child, move in child_move:
            child_h = cal_cost(child)
            if child_h < current_h:
                better_moves.append((child_h, child, move))
        if not better_moves:
            print("Đã bị kẹt ở đỉnh cục bộ! Thuật toán Stochastic dừng lại.")
            return current_state
        chosen_h, chosen_child, chosen_move = random.choice(better_moves)
        current_state = get_state(chosen_h, 0, chosen_child, current_state, chosen_move, step)
    return None

def do_RandomRestartHillClimbing(start_state, max_restarts=500):
    initial_h = cal_cost(start_state)
    for restart in range(max_restarts):
        current_state = get_state(initial_h, 0, start_state, None, "START", 0)
        step = 0 
        while step < 281000:
            node = current_state[2]
            current_h = current_state[0]
            if is_goal(node):
                print(f"Tuyệt vời! Đã tìm thấy đích sau {restart} lần Khởi động lại.")
                return current_state
            child_move = expand(node)
            step += 1
            better_moves = []
            for child, move in child_move:
                child_h = cal_cost(child)
                if child_h < current_h:
                    better_moves.append((child_h, child, move))
            if not better_moves:
                break
            chosen_h, chosen_child, chosen_move = random.choice(better_moves)
            current_state = get_state(chosen_h, 0, chosen_child, current_state, chosen_move, step)
    print(f"Đã đạt giới hạn {max_restarts} lần Khởi động lại nhưng vẫn bị kẹt.")
    return None

def do_LocalBeamSearch(start_state, k=3):
    initial_h = cal_cost(start_state)
    initial_state = get_state(initial_h, 0, start_state, None, "START", 0)
    beam = [initial_state]
    step = 0
    node_id_counter = 0
    reached = set()
    reached.add(to_tuple(start_state))
    
    while step < 281000:
        step += 1
        all_successors = []
        for current_state in beam:
            node = current_state[2]
            if is_goal(node):
                return current_state
            child_move = expand(node)
            for child, move in child_move:
                child_tuple = to_tuple(child)
                if child_tuple not in reached:
                    reached.add(child_tuple)
                    child_h = cal_cost(child)
                    node_id_counter += 1
                    s = get_state(child_h, node_id_counter, child, current_state, move, step)
                    all_successors.append(s)
        if not all_successors:
            print(f"Cả {k} nhánh đều bị kẹt! Thuật toán Local Beam Search dừng lại.")
            return beam[0]
        all_successors.sort(key=lambda x: (x[0], x[1]))
        beam = all_successors[:k]
    return None

def do_SimulatedAnnealing(start_state):
    current_h = cal_cost(start_state)
    current_state = get_state(current_h, 0, start_state, None, "START", 0)
    
    T = 100.0
    T_min = 1e-4
    alpha = 0.999
    step = 0
    
    while T > T_min and step < 281000:
        node = current_state[2]
        current_h = current_state[0]
        if is_goal(node):
            return current_state
        child_move = expand(node)
        step += 1
        
        child, move = random.choice(child_move)
        child_h = cal_cost(child)
        
        delta = child_h - current_h
        if delta < 0:
            current_state = get_state(child_h, 0, child, current_state, move, step)
        else:
            p = math.exp(-delta / T)
            if random.random() < p:
                current_state = get_state(child_h, 0, child, current_state, move, step)
        T *= alpha
    print("Đã giảm nhiệt độ đến mức tối thiểu nhưng chưa tìm thấy đích.")
    return None

BACKGROUND_COLOR = (0, 0, 0)
TILE_COLOR = (52, 152, 219) 
BTN_COLOR = (46, 204, 113)  
BACK_BTN_COLOR = (231, 76, 60) 
TEXT_COLOR = (255, 255, 255)      
LIGHT_TEXT = (220, 220, 220)       
LINE_COLOR = (80, 80, 80)          

TILE_SIZE = 100
MARGIN = 8
MATRIX_WIDTH = TILE_SIZE * 3 + MARGIN * 4 
GAME_WIDTH = MATRIX_WIDTH * 2 
LOG_PANEL_WIDTH = 550
TOTAL_WIDTH = GAME_WIDTH + LOG_PANEL_WIDTH
TOTAL_HEIGHT = 600

def get_grid_pos(index):
    return index // 3, index % 3

def get_pixel_coords(row, col, matrix_idx):
    offset_x = matrix_idx * MATRIX_WIDTH
    x = offset_x + MARGIN + col * (TILE_SIZE + MARGIN)
    y = MARGIN + row * (TILE_SIZE + MARGIN)
    return x, y

def draw_static_state(screen, font, flats, ignore_indices, solution_path=None, current_step_idx=0, scroll_line=None):
    screen.fill(BACKGROUND_COLOR)
    
    for m_idx, flat in enumerate(flats):
        for i, val in enumerate(flat):
            if val != 0 and (m_idx >= len(ignore_indices) or i != ignore_indices[m_idx]):
                row, col = get_grid_pos(i)
                x, y = get_pixel_coords(row, col, m_idx)
                pygame.draw.rect(screen, TILE_COLOR, (x, y, TILE_SIZE, TILE_SIZE), border_radius=8)
                text_surface = font.render(str(val), True, TEXT_COLOR)
                text_rect = text_surface.get_rect(center=(x + TILE_SIZE // 2, y + TILE_SIZE // 2))
                screen.blit(text_surface, text_rect)

    pygame.draw.line(screen, LINE_COLOR, (MATRIX_WIDTH, 0), (MATRIX_WIDTH, MATRIX_WIDTH), 3)

    btn_font = pygame.font.Font(None, 30)
    back_btn_rect = pygame.Rect(GAME_WIDTH - 160, TOTAL_HEIGHT - 60, 150, 45)
    pygame.draw.rect(screen, BACK_BTN_COLOR, back_btn_rect, border_radius=8)
    btn_surf = btn_font.render("Quay lai Menu", True, TEXT_COLOR)
    screen.blit(btn_surf, btn_surf.get_rect(center=back_btn_rect.center))
    
    pygame.draw.line(screen, LINE_COLOR, (GAME_WIDTH, 0), (GAME_WIDTH, TOTAL_HEIGHT), 3)
    
    log_font = pygame.font.Font(None, 24) 
    log_x = GAME_WIDTH + 15
    log_y = 15
    line_height = 20 
    
    title = pygame.font.Font(None, 35).render("LICH SU QUA TRINH", True, LIGHT_TEXT)
    screen.blit(title, (log_x, log_y))
    
    current_scroll = 0
    max_scroll = 0
    if solution_path is not None:
        all_lines = []
        for i in range(current_step_idx + 1):
            item = solution_path[i]
            r1 = item['matrix'][0]
            all_lines.append(f"--- Buoc {i} | Lenh: {item['move']} ---")
            
            if len(item['matrix']) > 1:
                r2 = item['matrix'][1]
                all_lines.append(f" M1: [{r1[0]}] [{r1[1]}] [{r1[2]}]  |  M2: [{r2[0]}] [{r2[1]}] [{r2[2]}]")
                all_lines.append(f"     [{r1[3]}] [{r1[4]}] [{r1[5]}]  |      [{r2[3]}] [{r2[4]}] [{r2[5]}]")
                all_lines.append(f"     [{r1[6]}] [{r1[7]}] [{r1[8]}]  |      [{r2[6]}] [{r2[7]}] [{r2[8]}]")
            else:
                all_lines.append(f" M1: [{r1[0]}] [{r1[1]}] [{r1[2]}]")
                all_lines.append(f"     [{r1[3]}] [{r1[4]}] [{r1[5]}]")
                all_lines.append(f"     [{r1[6]}] [{r1[7]}] [{r1[8]}]")
            all_lines.append("") 
            
        max_visible_lines = (TOTAL_HEIGHT - 60) // line_height
        max_scroll = max(0, len(all_lines) - max_visible_lines)
        if scroll_line is None:
            current_scroll = max_scroll 
        else:
            current_scroll = max(0, min(scroll_line, max_scroll))
            
        visible_lines = all_lines[current_scroll : current_scroll + max_visible_lines]
        for idx, line_text in enumerate(visible_lines):
            if line_text != "":
                surf = log_font.render(line_text, True, LIGHT_TEXT)
                screen.blit(surf, (log_x, log_y + 40 + idx * line_height))
                
    return back_btn_rect, current_scroll, max_scroll

def animate_transition(screen, font, clock, curr_item, nxt_item, solution_path, step_num, scroll_line, animation_steps=15):
    curr_flats = curr_item['matrix']
    nxt_flats = nxt_item['matrix']
    num_matrices = len(curr_flats)
    
    moving_info = []
    for i in range(num_matrices):
        c_flat = curr_flats[i]
        n_flat = nxt_flats[i]
        c_zero = c_flat.index(0)
        n_zero = n_flat.index(0)
        
        if c_zero != n_zero:
            moving_val = c_flat[n_zero]
            o_r, o_c = get_grid_pos(n_zero)
            n_r, n_c = get_grid_pos(c_zero)
            ox, oy = get_pixel_coords(o_r, o_c, i)
            nx, ny = get_pixel_coords(n_r, n_c, i)
            moving_info.append({
                'idx': i, 'val': moving_val,
                'ox': ox, 'oy': oy, 'nx': nx, 'ny': ny,
                'ignore_idx': n_zero
            })
        else:
            moving_info.append(None)
            
    for step in range(animation_steps + 1):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return False
        progress = step / animation_steps
        
        ignore_indices = [info['ignore_idx'] if info else -1 for info in moving_info]
        
        draw_static_state(screen, font, curr_item['matrix'], ignore_indices, solution_path, step_num, scroll_line)
        
        for info in moving_info:
            if info:
                cx = info['ox'] + (info['nx'] - info['ox']) * progress
                cy = info['oy'] + (info['ny'] - info['oy']) * progress
                pygame.draw.rect(screen, TILE_COLOR, (cx, cy, TILE_SIZE, TILE_SIZE), border_radius=8)
                ts = font.render(str(info['val']), True, TEXT_COLOR)
                screen.blit(ts, ts.get_rect(center=(cx + TILE_SIZE // 2, cy + TILE_SIZE // 2)))
                
        pygame.display.flip()
        clock.tick(60)
    return True

def show_config_menu(screen):
    title_font = pygame.font.Font(None, 45)
    btn_font = pygame.font.Font(None, 30)
    btn_width = 500
    btn_height = 55
    center_x = TOTAL_WIDTH // 2 - btn_width // 2
    
    btns = [
        pygame.Rect(center_x, 150, btn_width, btn_height),
        pygame.Rect(center_x, 230, btn_width, btn_height),
        pygame.Rect(center_x, 310, btn_width, btn_height)
    ]
    labels = [
        "1. Khuyet Start (Chay 2 ma tran vao 1 Goal)",
        "2. Khuyet Goal (Chay 1 ma tran vao 2 Goal)",
        "3. Khuyet ca hai (Chay 2 ma tran vao 2 Goal)"
    ]
    returns = ["NO_START", "NO_GOAL", "NO_BOTH"]
    
    while True:
        screen.fill(BACKGROUND_COLOR)
        title_surf = title_font.render("CAU HINH BAO TOAN (KHUYET 2 O)", True, LIGHT_TEXT)
        screen.blit(title_surf, title_surf.get_rect(center=(TOTAL_WIDTH // 2, 70)))
        
        for i, rect in enumerate(btns):
            pygame.draw.rect(screen, BTN_COLOR, rect, border_radius=10)
            btn_surf = btn_font.render(labels[i], True, TEXT_COLOR)
            screen.blit(btn_surf, btn_surf.get_rect(center=rect.center))
            
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return None
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                for i, rect in enumerate(btns):
                    if rect.collidepoint(event.pos):
                        return returns[i]
        pygame.display.flip()

def show_algorithm_menu(screen):
    title_font = pygame.font.Font(None, 40)
    btn_font = pygame.font.Font(None, 24)
    btn_width = 320
    btn_height = 45
    
    col1_x = TOTAL_WIDTH // 4 - btn_width // 2
    col2_x = 3 * TOTAL_WIDTH // 4 - btn_width // 2
    
    btns = []
    labels = [
        "1. BFS", "2. DFS", "3. DLS", "4. UCS",
        "5. Greedy", "6. A*", "7. IDA*", "8. Simple Hill",
        "9. Steepest-Ascent", "10. Stochastic Hill", 
        "11. Random-Restart", "12. Local Beam", "13. Simulated Anneal"
    ]
    returns = [
        "BFS", "DFS", "DLS", "UCS",
        "GREEDY", "ASTAR", "IDASTAR", "SIMPLE_HILL",
        "STEEPEST_HILL", "STOCHASTIC_HILL", 
        "RANDOM_RESTART_HILL", "LOCAL_BEAM", "SIMULATED_ANNEALING"
    ]
    
    for i in range(13):
        col = col1_x if i < 7 else col2_x
        row = i if i < 7 else i - 7
        y = 100 + row * 60
        rect = pygame.Rect(col, y, btn_width, btn_height)
        btns.append(rect)
        
    while True:
        screen.fill(BACKGROUND_COLOR)
        title_surf = title_font.render("CHON THUAT TOAN TIM KIEM", True, LIGHT_TEXT)
        screen.blit(title_surf, title_surf.get_rect(center=(TOTAL_WIDTH // 2, 40)))
        
        for i, rect in enumerate(btns):
            pygame.draw.rect(screen, BTN_COLOR, rect, border_radius=10)
            btn_surf = btn_font.render(labels[i], True, TEXT_COLOR)
            screen.blit(btn_surf, btn_surf.get_rect(center=rect.center))
            
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return None
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                for i, rect in enumerate(btns):
                    if rect.collidepoint(event.pos):
                        return returns[i]
        pygame.display.flip()

def show_loading_screen(screen):
    screen.fill(BACKGROUND_COLOR)
    font = pygame.font.Font(None, 40)
    text_surf = font.render("Dang tinh toan... (Co the mat nhieu thoi gian)", True, LIGHT_TEXT)
    text_rect = text_surf.get_rect(center=(TOTAL_WIDTH // 2, TOTAL_HEIGHT // 2))
    screen.blit(text_surf, text_rect)
    pygame.display.flip()

def run_visualizer(screen, solution_path):
    font = pygame.font.Font(None, 50)
    clock = pygame.time.Clock()
    current_step_idx = 0
    curr_item = solution_path[current_step_idx]
    scroll_line = None
    auto_scroll = True
    max_scroll = 0
    
    back_btn_rect, scroll_line_out, max_scroll = draw_static_state(
        screen, font, curr_item['matrix'], [],
        solution_path=solution_path, current_step_idx=0, scroll_line=scroll_line
    )
    pygame.display.flip()
    time.sleep(1.5) 
    running = True
    is_finished = False
    
    while running:
        if is_finished:
            back_btn_rect, scroll_line_out, max_scroll = draw_static_state(
                screen, font, solution_path[-1]['matrix'], [],
                solution_path=solution_path, current_step_idx=len(solution_path)-1,
                scroll_line=scroll_line
            )
            pygame.display.flip()
            clock.tick(30)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return False
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                if back_btn_rect.collidepoint(event.pos):
                    return True
            if event.type == pygame.MOUSEWHEEL:
                if scroll_line is None:
                    scroll_line = max_scroll
                scroll_line -= event.y * 3 
                auto_scroll = False
                
        if not is_finished and current_step_idx < len(solution_path) - 1:
            curr_item = solution_path[current_step_idx]
            nxt_item = solution_path[current_step_idx + 1]
            step_number = current_step_idx + 1
            if auto_scroll:
                scroll_line = None
            if not animate_transition(screen, font, clock, curr_item, nxt_item, solution_path, step_number, scroll_line):
                return False
            current_step_idx += 1
            time.sleep(0.3) 
        elif not is_finished:
            is_finished = True
    return True

if __name__ == "__main__":
    pygame.init()
    screen = pygame.display.set_mode((TOTAL_WIDTH, TOTAL_HEIGHT))
    pygame.display.set_caption("Multi-Agent 8-Puzzle Search (2 Matrices)")
    
    while True:
        config_choice = show_config_menu(screen)
        if config_choice is None:
            break

        selected_algorithm = show_algorithm_menu(screen)
        if selected_algorithm is None:
            break
            
        show_loading_screen(screen)
        
        base_start = [[1, 2, 3], [4, -1, 5], [7, -1, 6]]
        base_goal = [[1, 2, 3], [4, 5, 6], [7, -1, -1]]
        
        full_start_1 = [[1, 2, 3], [4, 0, 5], [7, 8, 6]]
        full_goal_1 = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

        if config_choice == "NO_START":
            start_state = generate_2_states(base_start)
            GOAL_MATRICES = [full_goal_1]
        elif config_choice == "NO_GOAL":
            start_state = [full_start_1]
            GOAL_MATRICES = generate_2_states(base_goal)
        elif config_choice == "NO_BOTH":
            start_state = generate_2_states(base_start)
            GOAL_MATRICES = generate_2_states(base_goal)
            
        final_node = None

        if selected_algorithm == "BFS":
            frontier = queue.Queue()
            reached = set() 
            frontier_set = set()
            initial_node = get_state(0, 0, start_state, None, None, 0)
            frontier.put(initial_node)
            frontier_set.add(to_tuple(start_state))
            final_node = do_BFS(reached, frontier, frontier_set)
            
        elif selected_algorithm == "DFS":
            frontier = []
            reached = set() 
            frontier_set = set()
            initial_node = get_state(0, 0, start_state, None, None, 0)
            frontier.append(initial_node)
            frontier_set.add(to_tuple(start_state))
            final_node = do_DFS(reached, frontier, frontier_set)
            
        elif selected_algorithm == "DLS":
            for depth in range(100):
                frontier = []
                reached = set() 
                frontier_set = set()
                initial_node = get_state(0, 0, start_state, None, None, 0)
                frontier.append(initial_node)
                frontier_set.add(to_tuple(start_state))
                final_node = do_DLS(reached, frontier, frontier_set, limit_depth=depth)
                if final_node is not None:
                    break
                    
        elif selected_algorithm == "UCS":
            frontier = queue.PriorityQueue()
            reached = {} 
            initial_node = get_state(0, 0, start_state, None, None, 0)
            frontier.put(initial_node)
            reached[to_tuple(start_state)] = 0
            final_node = do_UCS(reached, frontier)

        elif selected_algorithm == "GREEDY":
            frontier = queue.PriorityQueue()
            reached = {} 
            initial_node = get_state(0, 0, start_state, None, None, 0)
            frontier.put(initial_node)
            reached[to_tuple(start_state)] = 0
            final_node = do_Greedy(reached, frontier)

        elif selected_algorithm == "ASTAR":
            frontier = queue.PriorityQueue()
            reached = {} 
            initial_node = get_state(0, 0, start_state, None, None, 0)
            frontier.put(initial_node)
            reached[to_tuple(start_state)] = 0
            final_node = do_AStar(reached, frontier)

        elif selected_algorithm == "IDASTAR":
            start_tuple = to_tuple(start_state)
            limit_f = cal_cost(start_state)
            while limit_f < float('inf'):
                frontier = []
                reached = {}
                initial_node = get_state(0, 0, start_state, None, None, 0)
                frontier.append(initial_node)
                final_node, next_limit = do_IDAStar(reached, frontier, limit_f)
                if final_node is not None:
                    break
                if next_limit == float('inf'):
                    break
                limit_f = next_limit

        elif selected_algorithm == "SIMPLE_HILL":
            final_node = do_SimpleHillClimbing(start_state)

        elif selected_algorithm == "STEEPEST_HILL":
            final_node = do_SteepestAscent(start_state)
            
        elif selected_algorithm == "STOCHASTIC_HILL":
            final_node = do_StochasticHillClimbing(start_state)
            
        elif selected_algorithm == "RANDOM_RESTART_HILL":
            final_node = do_RandomRestartHillClimbing(start_state, max_restarts=500)
            
        elif selected_algorithm == "LOCAL_BEAM":
            final_node = do_LocalBeamSearch(start_state, k=3)

        elif selected_algorithm == "SIMULATED_ANNEALING":
            final_node = do_SimulatedAnnealing(start_state)
        
        if final_node is not None:
            solution_path = extract_path(final_node)
            keep_playing = run_visualizer(screen, solution_path) 
            if not keep_playing:
                break
        else:
            screen.fill(BACKGROUND_COLOR)
            font = pygame.font.Font(None, 40)
            text_surf = font.render("Khong the giai hoac qua lau (Vuot limit)!", True, BACK_BTN_COLOR)
            text_rect = text_surf.get_rect(center=(TOTAL_WIDTH // 2, TOTAL_HEIGHT // 2))
            screen.blit(text_surf, text_rect)
            pygame.display.flip()
            time.sleep(3)
            
    pygame.quit()
    print("Da thoat chuong trinh an toan.")
```
