```python
import numpy as np
import random
```

```python
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

directions = {
    "UP": (-1, 0),
    "DOWN": (1, 0),
    "LEFT": (0, -1),
    "RIGHT": (0, 1)
}

xdirections = {
    "UP": "DOWN",
    "DOWN": "UP",
    "LEFT": "RIGHT",
    "RIGHT": "LEFT"
}
```

```python
def interpret_input():
    floor = np.random.randint(0, 2, size=(rows, cols))
    robot_pos = (random.randint(0, rows-1), random.randint(0, cols-1))
    print(f"Robot Position: {robot_pos}, Floor State:\n{floor}\n")
    
    return robot_pos, floor
```

```python
def rule_match(robot_pos, floor):
    r, c = robot_pos
    if floor[r, c] == 1:
        return "CLEAN"
    
    pos_moves = []
    if r > 0:
        pos_moves.append("UP")
    if r < rows - 1:
        pos_moves.append("DOWN")
    if c > 0:
        pos_moves.append("LEFT")
    if c < cols - 1:
        pos_moves.append("RIGHT")
    return pos_moves
```

```python
max_battery = rows * cols * 10
battery = max_battery
robot_pos, floor = interpret_input()
last_move = None

while battery > 0:
    possible_actions = rule_match(robot_pos, floor)
    action = (possible_actions if isinstance(possible_actions, str) 
    else random.choice([a for a in possible_actions if a != xdirections.get(last_move, "")]))
    r, c = robot_pos

    if action == "CLEAN":
        floor[r, c] = 0
    else:
        move = action
        dr, dc = directions[move]
        robot_pos = (r + dr, c + dc)
        last_move = action
    
    print(f"Robot Position: {robot_pos}, Steps: {(max_battery - battery) + 1}, Action: {action}\n")
    
    if np.sum(floor) == 0:
        print("All spots are clean!")
        break
    
    print(f"Current floor state:\n{floor}\n")
    battery -= 1
    print(f"Battery remaining: {battery/max_battery*100:.2f}%\n")
```
